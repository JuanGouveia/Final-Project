```python
import pandas as pd
import numpy as np
```

```python
races = pd.read_csv('../clean_data/races_clean.csv')
circuits = pd.read_csv('../clean_data/circuits_clean.csv')
```

```python
races.head()
```

```python
circuits.head()
```

```python
calendar = pd.merge(races, circuits, on='circuitId', how='left')

calendar['location'] = calendar['location'] + ', ' + calendar['country']

calendar.head()
```

```python
calendar = calendar.drop(['Unnamed: 0_x', 'raceId', 'circuitId', 'Unnamed: 0_y', 'circuitRef', 'country', 'lat', 'lng', 'alt'], axis=1)

calendar.head()
```

```python
calendar.isna().sum()
```

```python
calendar.shape[0] == races.shape[0]
```

```python
calendar = calendar.rename(columns={'round': 'Round', 'name_x': 'Race', 'date': 'Date', 'name_y': 'Circuit', 'location': 'Location'})

calendar.head()
```

```python
calendar.dtypes
```

```python
calendar['Date'] = calendar['Date'].astype('datetime64[ns]')

calendar.dtypes
```

```python
cal = calendar.to_csv('../streamlit_data/calendar.csv')
```
