```python
import pandas as pd
import numpy as np
```

```python
races = pd.read_csv('../clean_data/races_clean.csv')
results = pd.read_csv('../clean_data/results_clean.csv')
circuits = pd.read_csv('../clean_data/circuits_clean.csv')
quali = pd.read_csv('../clean_data/qualifying_clean.csv')
drivers = pd.read_csv('../clean_data/drivers_clean.csv')
const = pd.read_csv('../clean_data/constructor_clean.csv')
status = pd.read_csv('../clean_data/status_clean.csv')
```

# Análisis del dataframe

---

Procedemos a transformar nuestro dataframe según nuestros objetivos, eliminando ciertas columnas y modificando ciertos parámetros

```python
results.shape
```

```python
results.head()
```

```python
results = results.drop(['position', 'milliseconds', 'fastestLap', 'fastestLapTime', 'fastestLapSpeed'], axis=1)

results.rename(columns={'number': 'Driver Number', 'grid': 'Grid Position', 'positionOrder': 'Position', 'points': 'Points', 'laps': 'Laps',
                        'time': 'Time'}, inplace=True)
                                                            # eliminamos las columnas que no necesitemos y renombramos las que si necesitamos
results.head()
```

```python
races.shape
```

```python
races.head()
```

```python
races = races.drop('date', axis=1)

races.rename(columns={'year': 'Seasons', 'round': 'Round', 'name': 'Races'}, inplace=True)

races.head()
```

```python
circuits.shape
```

```python
circuits.head()
```

```python
circuits['Location'] = circuits['location'] + ', ' + circuits['country']        # convertimos las columnas 'location' y 'country' en una sola

circuits = circuits.drop(['circuitRef', 'location', 'country', 'lat', 'lng', 'alt'], axis=1)

circuits.rename(columns={'name': 'Circuit'}, inplace=True)

circuits.head()
```

```python
quali.head()
```

```python
quali = quali.drop('number', axis=1)

quali.rename(columns={'position': 'Qualify Position', 'q1': 'Q1', 'q2': 'Q2', 'q3': 'Q3'}, inplace=True)

quali.head()
```

```python
drivers.shape
```

```python
drivers.head()
```

```python
drivers['Driver'] = drivers['forename'] + ' ' + drivers['surname']      # convertimos las columnas 'forename' y 'surname' en una sola

drivers = drivers.drop(['driverRef', 'forename', 'surname', 'dob', 'nationality'], axis=1)

drivers.head()
```

```python
const.shape
```

```python
const.head()
```

```python
const = const.drop(['constructorRef', 'nationality'], axis=1)

const.rename(columns={'name': 'Team'}, inplace=True)

const.head()
```

```python
status.shape
```

```python
status.head()
```

```python
status.rename(columns={'status': 'Status'}, inplace=True)

status.head()
```

# Merge

---

Procedemos a realizar los distintos merges para convertir todos los dataframes en uno solo

```python
status.head()
```

```python
results.head()
```

```python
res_stat = results.merge(status, on='statusId', how='left')

res_stat.head()
```

```python
const.head()
```

```python
res_stat_const = res_stat.merge(const, on='constructorId', how='left')

res_stat_const.head()
```

```python
drivers.head()
```

```python
res_stat_const_driv = res_stat_const.merge(drivers, on='driverId', how='left')

res_stat_const_driv.head()
```

```python
quali.head()
```

```python
res_stat_const_driv_quali = res_stat_const_driv.merge(quali, on=['driverId', 'raceId', 'constructorId'], how='left')

res_stat_const_driv_quali.head()
```

```python
races.head()
```

```python
circuits.head()
```

```python
circ_race = circuits.merge(races, on='circuitId', how='left')

circ_race.head()
```

```python
pd.set_option('display.max_columns', 50)

results = res_stat_const_driv_quali.merge(circ_race, on='raceId', how='left')

results.head()
```

```python
results.shape
```

```python
results.isna().sum()
```

```python
null_values = results[results['qualifyId'].isnull() & results['Qualify Position'].isnull() & results['Q1'].isnull() & results['Q2'].isnull() & results['Q3'].isnull()]['Seasons']
                                        # filtramos por los valores nulos que tenemos
null_values.value_counts().head()
```

```python
results = results.dropna()                      # eliminamos las filas que contengan valores nulos debido a que pertenecen a temporadas en las cuales no
results = results.reset_index(drop=True)        # disponemos de esos datos

results.tail()
```
