```python
import pandas as pd
import numpy as np
```

```python
circuits = pd.read_csv('../data/circuits.csv')
constructors = pd.read_csv('../data/constructors.csv')
drivers = pd.read_csv('../data/drivers.csv')
races = pd.read_csv('../data/races.csv')
results = pd.read_csv('../data/results.csv')
d_points = pd.read_csv('../data/driver_points_es.csv')
c_points = pd.read_csv('../data/constructor_points_es.csv')
weather = pd.read_csv('../data/weather_es.csv')
safety = pd.read_csv('../data/safety_car.csv')
```

# Análisis del dataframe

---

Procedemos a acomodar los distintos dataframes, filtrando por los grandes premios de España y acomodando los dataframes según nuestros intereses

```python
circuits.shape
```

```python
circuits.head()
```

```python
circuits['name'].unique()
```

```python
circuit_es = circuits[circuits['name'] == 'Circuit de Barcelona-Catalunya']     # filtramos por el Circuit de Barcelona-Catalunya
circuit_es = circuit_es.drop('url', axis=1)     # borramos la columna 'url' debido a que no nos interesa
circuit_es = circuit_es.sort_values(by='circuitId', ascending=False)

circuit_es
```

```python
races.shape
```

```python
races.head()
```

```python
races['name'].value_counts().head(10)       # buscamos el gran premio de España
```

```python
races_es = races[(races['name'] == 'Spanish Grand Prix') & (races['year'].between(1996, 2022))]     # filtramos por los grandes premios de España a partir del año en el cuál empezamos a tener registros fiables de safety cars
races_es = races_es.drop(['fp1_date', 'fp1_time', 'fp2_date', 'fp2_time', 'fp3_date', 'fp3_time', 'quali_date', 'quali_time', 'sprint_date', 'sprint_time', 'url'], axis=1)        # eliminamos las siguientes columnas por falta de datos
races_es = races_es.sort_values(by='raceId', ascending=False)

races_es.head()
```

```python
races_es['time'].value_counts()
```

```python
races_es['time'] = races_es['time'].replace('\\N', '14:00:00')      # reemplazamos los valores faltantes con los datos que averiguamos

races_es['time'].value_counts()
```

```python
races_es.head()
```

```python
races_es.shape
```

```python
results.shape
```

```python
results.head()
```

```python
results_es = results[results['raceId'].isin(races_es['raceId'])]
results_es = results_es.drop(['number', 'position', 'positionText', 'positionOrder', 'points', 'laps', 'time', 'milliseconds', 'fastestLap', 'rank', 'fastestLapTime', 'fastestLapSpeed', 'statusId'], axis=1)
results_es = results_es.sort_values(by='raceId', ascending=False)       # nos quedamos unicamente con la posición de parrila debido a que lo demás son datos que no tendremos

len(results_es['raceId'].unique()) == len(races_es['raceId'].unique())      # chequeamos las longitudes de la columna 'raceId' en ambos dataframes para asegurarnos de que esten las mismas carreras
```

```python
results_es.head()
```

```python
results_es.shape
```

```python
drivers.shape
```

```python
drivers.head()
```

```python
drivers_es = drivers[drivers['driverId'].isin(results_es['driverId'])]        # filtramos por el driverId de el dataframe ya filtrado de los grandes premios de España   
drivers_es = drivers_es.drop(['number', 'code', 'url'], axis=1)
drivers_es = drivers_es.sort_values(by='driverId', ascending=False)

drivers_es.head()
```

```python
drivers_es.shape
```

```python
constructors.shape
```

```python
constructors.head()
```

```python
constructors_es = constructors[constructors['constructorId'].isin(results_es['constructorId'])]
constructors_es = constructors_es.drop('url', axis=1)
constructors_es.rename(columns={'name': 'constructor_name'}, inplace=True)
constructors_es = constructors_es.drop(constructors_es[constructors_es['constructor_name'] == 'forti'].index)       # eliminamos la escudería 'forti', ya que no figura en nuestros datos de escuderias
constructors_es = constructors_es.sort_values(by='constructorId', ascending=False)

constructors_es.head()
```

# Merge

---

Procedemos a realizar los distintos merges para convertir todos los dataframes en uno solo

```python
races_es.head()
```

```python
circuit_es
```

```python
race_circ_es = circuit_es.merge(races_es, on='circuitId', how='left')

race_circ_es.head()
```

```python
results_es.head()
```

```python
race_circ_res_es = race_circ_es.merge(results_es, on='raceId', how='left')

race_circ_res_es.head()
```

```python
drivers_es.head()
```

```python
race_circ_res_driv_es = race_circ_res_es.merge(drivers_es, on='driverId', how='left')

race_circ_res_driv_es.head()
```

```python
constructors_es.head()
```

```python
race_circ_res_driv_cons_es = race_circ_res_driv_es.merge(constructors_es, on='constructorId', how='left')

race_circ_res_driv_cons_es.head()
```

```python
d_points.head()
```

```python
race_circ_res_driv_cons_dpts_es = race_circ_res_driv_cons_es.merge(d_points, on=['driverRef', 'year'], how='left')

race_circ_res_driv_cons_dpts_es.head()
```

```python
c_points.head()
```

```python
race_circ_res_driv_cons_dpts_cpts_es = race_circ_res_driv_cons_dpts_es.merge(c_points, on=['constructorRef', 'year'], how='left')

race_circ_res_driv_cons_dpts_cpts_es.head()
```

```python
weather.head()
```

```python
pd.set_option('display.max_columns', 40)

sc_record_temp = race_circ_res_driv_cons_dpts_cpts_es.merge(weather, on='year', how='left')

sc_record_temp.head()
```

```python
sc_record_temp.isna().sum()
```

```python
sc_record_temp = sc_record_temp.dropna()        # eliminamos las filas que contengan nulos debido a que son correspondientes a pilotos y escuderias que no puntuaron y por eso no coinciden con nuestros otros registros, genereando así, valores nulos
```

```python
sc_record_temp.isna().sum()
```

```python
sc_record_temp.shape
```

```python
sc_record_temp.columns
```

```python
sc_record_temp = sc_record_temp.rename(columns={
'name_x': 'name',
'name_y': 'grand_prix',
'nationality_x': 'driver_nationality',
'nationality_y': 'const_nationality'
})

sc_record_temp.head()
```

# Columnas extras

---

Procedemos a crear nuevas columnas con datos extras que obtuvimos

```python
safety.shape
```

```python
safety.head()
```

```python
safety_es = safety[safety['Grand Prix'] == 'Spanish Grand Prix']        # filtramos los safety cars que han habido por el gran premio de España
safety_es.rename(columns={'Grand Prix': 'grand_prix', 'Cause': 'cause', 'Deployed (lap)': 'deployed_lap', 'Called in (lap)': 'called_in_lap', 'Number of full laps': 'number_full_laps', 'Year': 'year'}, inplace=True)

safety_es.head()
```

```python
sc_record_temp['safety_car'] = ((sc_record_temp['year'] == 2021) & (sc_record_temp['lap'].isin([8, 9, 10]))) | ((sc_record_temp['year'] == 2019) & (sc_record_temp['lap'].isin([46, 47, 48, 49, 50, 51, 52])) | (sc_record_temp['year'] == 2018) & (sc_record_temp['lap'].isin([1, 2, 3, 4, 5, 6])) | (sc_record_temp['year'] == 2009) & (sc_record_temp['lap'].isin([1, 2, 3, 4, 5])) | (sc_record_temp['year'] == 2008) & (sc_record_temp['lap'].isin([1, 2, 3, 4])))

sc_record_temp.head()       # definimos que rellene las filas de la columna 'safety_car' con 'True' o 'False' según el registro que tenmos de safety cars
```

```python
sc_record_temp['safety_car'].value_counts()
```

```python
sc_record_temp['turns'] = 16        # las curvas que tiene el circuito

sc_record_temp.head()
```

```python
length_values = {1996: '4727', 1997: '4728', 1998: '4728', 1999: '4728', 2000: '4730', 2001: '4730', 2002: '4730', 2003: '4730', 2004: '4627',
                 2005: '4627', 2006: '4627', 2007: '4655', 2008: '4655', 2009: '4655', 2010: '4655', 2011: '4655', 2012: '4655', 2013: '4655',
                 2014: '4655', 2015: '4655', 2016: '4655', 2017: '4655', 2018: '4655', 2019: '4655', 2020: '4655', 2021: '4655', 2022: '4675'
                }

sc_record_temp['length'] = ''       # la longitud en km del circuito a lo largo de los años

for year in sc_record_temp['year'].unique():
    mask = sc_record_temp['year'] == year       # definimos que rellene las filas con la longitud predefinida de acuerdo al año correspondiente
    sc_record_temp.loc[mask, 'length'] = length_values[year]

sc_record_temp.head()
```
