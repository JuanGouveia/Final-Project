# Web scraping

Iniciamos el web scraping mediante selenium

```python
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.common.by import By # By es para buscar por tag, clase, id...

import time

import warnings
warnings.filterwarnings('ignore')
```

```python
#driver configuration
opciones = Options()

opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.headless=False    # si True, no aperece la ventana (headless=no visible)
#opciones.add_argument('--start-maximized')         # comienza maximizado
#opciones.add_argument('user-data-dir=selenium')    # mantiene las cookies
#opciones.add_extension('driver_folder/adblock.crx')       # adblocker
#opciones.add_argument('--incognito')
```

```python
PATH = ChromeDriverManager().install()

driver = webdriver.Chrome(PATH)

driver.get('https://f1.fandom.com/wiki/Safety_Car')
```

```python
aceptar = driver.find_element(By.CSS_SELECTOR, 'body > div:nth-child(19) > div > div > div._1r08nyekFdI7_2d8r3AIBf > div._2O--J403t2VqCuF8XJAZLK')

time.sleep(3)

aceptar.click()
```

### Extracción de información

---

Procedemos a encontrar la tabla que deseamos y la extraemos, convirtiéndola en un dataframe para poder manipularla con pandas

```python
tabla = driver.find_elements(By.TAG_NAME, 'table')
len(tabla)
```

```python
print(tabla[1].text)
```

```python
filas = tabla[1].find_elements(By.TAG_NAME, 'tr')
print(filas[0].text)
```

```python
data = []


for f in filas:
    
    elems = f.find_elements(By.TAG_NAME, 'td')
    
    tmp = []
    
    for e in elems:
        tmp.append(e.text)
        
    data.append(tmp)
    
    
print(data)
```

```python
Safety_Car = pd.DataFrame(data)

Safety_Car.head(15)
```

```python
driver.quit()
```

```python
new_col = ['Grand Prix', 'Cause', 'Deployed (lap)', 'Called in (lap)', 'Number of full laps']       # renombramos las columnas
Safety_Car.columns = new_col

Safety_Car.head(15)
```

```python
Safety_Car = Safety_Car.drop(0).reset_index(drop=True)
```

### Detalles

---

Debido a la estructura de la tabla en la página web, hay algunas irregularidades en el dataframe, así que procedemos a arreglarlo

```python
for i in range(len(Safety_Car['Grand Prix'])):
    if Safety_Car['Grand Prix'][i] is not None and 'Grand Prix' not in Safety_Car['Grand Prix'][i]:
        Safety_Car['Number of full laps'][i] = Safety_Car['Called in (lap)'][i]
        Safety_Car['Called in (lap)'][i] = Safety_Car['Deployed (lap)'][i]
        Safety_Car['Deployed (lap)'][i] = Safety_Car['Cause'][i]
        Safety_Car['Cause'][i] = Safety_Car['Grand Prix'][i]
        Safety_Car['Grand Prix'][i] = Safety_Car['Grand Prix'][i-1]
```

```python
fila_297 = Safety_Car.loc[297]

fila_297['Grand Prix'] = '2022 Emilia Romagna Sprint'
fila_297['Cause'] = 'Accident (Zhou)'
fila_297['Deployed (lap)'] = 1
fila_297['Called in (lap)'] = 4
fila_297['Number of full laps'] = 2

Safety_Car.loc[297] = fila_297
```

```python
Safety_Car['Year'] = Safety_Car['Grand Prix'].str.extract('(\d{4})')        # creamos una nueva columna con las fechas usando regex
Safety_Car.head()
```

```python
Safety_Car['Grand Prix'] = Safety_Car['Grand Prix'].str.replace(r'^\d{4}\s', '', regex=True)        # eliminamos las fechas con regex
Safety_Car.head()
```

```python
Safety_Car['Grand Prix'].unique()
```

```python
Safety_Car = Safety_Car.sort_values(by = 'Year', ascending = False)     # ordenamos el dataframe con respecto a las fechas más recientes
Safety_Car = Safety_Car.reset_index(drop = True)

Safety_Car
```
